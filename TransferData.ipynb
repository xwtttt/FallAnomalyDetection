```python
import pandas as pd
import numpy as np
import os
```

```python
import matplotlib.pyplot as plt
```

```python
# 获取文件列表
def get_data_path(dir_name):
    data_path = []
    dir_list = os.listdir('./UseDataset/' + dir_name)
    if 'desktop.ini' in dir_list:
        dir_list.remove('desktop.ini')
    elif 'images.rar' in dir_list:
        dir_list.remove('images.rar')
    for i in range(len(dir_list)):
        data_path.append(os.path.join(os.getcwd(), 'UseDataset', dir_name, dir_list[i]))
    return data_path
```

```python
data_path_SA01 = get_data_path('SA01')
```

```python
# 读取不同的数据
Walking_slowly_data = pd.read_csv(data_path_SA01[0],
                                    names=['Acc_X_AD', 'Acc_Y_AD', 'Acc_Z_AD', 'Rot_X', 'Rot_Y', 'Rot_Z', 'Acc_X_MM', 'Acc_Y_MM', 'Acc_Z_MM'])
Walking_quickly_data = pd.read_csv(data_path_SA01[1],
                                    names=['Acc_X_AD', 'Acc_Y_AD', 'Acc_Z_AD', 'Rot_X', 'Rot_Y', 'Rot_Z', 'Acc_X_MM', 'Acc_Y_MM', 'Acc_Z_MM'])
Jogging_slowly_data = pd.read_csv(data_path_SA01[2],
                                    names=['Acc_X_AD', 'Acc_Y_AD', 'Acc_Z_AD', 'Rot_X', 'Rot_Y', 'Rot_Z', 'Acc_X_MM', 'Acc_Y_MM', 'Acc_Z_MM'])
Jogging_quickly_data = pd.read_csv(data_path_SA01[3],
                                    names=['Acc_X_AD', 'Acc_Y_AD', 'Acc_Z_AD', 'Rot_X', 'Rot_Y', 'Rot_Z', 'Acc_X_MM', 'Acc_Y_MM', 'Acc_Z_MM'])
```

```python
# 删除不需要的列
Walking_slowly_data = Walking_slowly_data.drop(columns=['Acc_X_MM', 'Acc_Y_MM', 'Acc_Z_MM'])
Walking_quickly_data = Walking_quickly_data.drop(columns=['Acc_X_MM', 'Acc_Y_MM', 'Acc_Z_MM'])
Jogging_slowly_data = Jogging_slowly_data.drop(columns=['Acc_X_MM', 'Acc_Y_MM', 'Acc_Z_MM'])
Jogging_quickly_data = Jogging_quickly_data.drop(columns=['Acc_X_MM', 'Acc_Y_MM', 'Acc_Z_MM'])
```

```python
Walking_slowly_data2 = (Walking_slowly_data - Walking_slowly_data.mean() ) / Walking_slowly_data.std()
```

```python
Walking_quickly_data2 = (Walking_quickly_data - Walking_quickly_data.mean()) / Walking_quickly_data.std()
```

```python
Walking_data = pd.concat([Walking_slowly_data2, Walking_quickly_data2])
```

```python
Walking_data.to_csv('fd/Walking_data.csv')
```
